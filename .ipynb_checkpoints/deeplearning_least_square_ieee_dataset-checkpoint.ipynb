```python
import numpy as np
import pandas as pd
import math
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optimizers
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torch.utils.data import random_split
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

#　ランダムシードの設定
if __name__ == '__main__':
    np.random.seed(0)
    torch.manual_seed(0)
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    
'''クラス'''

# データセット作成用クラス
class MyDataset(torch.utils.data.Dataset):
    def __init__(self, df, features, labels):
        self.features_values = df[features].values
        self.labels = df[labels].values

    # len()を使用すると呼ばれる
    def __len__(self):
        return len(self.features_values)
    
    # 要素を参照すると呼ばれる関数    
    def __getitem__(self, idx):
        features_x = torch.FloatTensor(self.features_values[idx])
        labels = torch.FloatTensor(self.labels[idx])
        return features_x, labels
    
    
# ネットワークの定義
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # 全結合層
        self.l1 = nn.Linear(4, 1000)
        self.a1 = nn.ReLU()
        self.bn1 = nn.BatchNorm1d(1000)
        self.l2 = nn.Linear(1000, 1000)
        self.a2 = nn.ReLU()
        self.bn2 = nn.BatchNorm1d(1000)
        self.l3 = nn.Linear(1000, 4)
        self.a3 = nn.Sigmoid()
        
    def forward(self, aoa):
        # 全結合層
        x = self.l1(aoa)
        x = self.a1(x)
        x = self.bn1(x)
        x = self.l2(x)
        x = self.a2(x)
        x = self.bn2(x)
        x = self.l3(x)
        c_val = self.a3(x)
        
        # 受信機位置のベクトル化
        pos_rx1 = torch.tensor(POS_RX[0]).view(-1, 1)
        pos_rx2 = torch.tensor(POS_RX[1]).view(-1, 1)
        pos_rx3 = torch.tensor(POS_RX[2]).view(-1, 1)
        pos_rx4 = torch.tensor(POS_RX[3]).view(-1, 1)

        # 最小二乗法
        I = torch.eye(2) # 単位行列
        
        pos_list = []

        for i in np.arange(len(aoa)):
            # 初期化
            R = 0
            q = 0

            # 方向ベクトル（AoA）
            n1 = torch.tensor([math.cos(math.radians(aoa[i][0])), math.sin(math.radians(aoa[i][0]))]).view(-1, 1)
            n2 = torch.tensor([math.cos(math.radians(aoa[i][1])), math.sin(math.radians(aoa[i][1]))]).view(-1, 1)
            n3 = torch.tensor([math.cos(math.radians(aoa[i][2])), math.sin(math.radians(aoa[i][2]))]).view(-1, 1)
            n4 = torch.tensor([math.cos(math.radians(aoa[i][3])), math.sin(math.radians(aoa[i][3]))]).view(-1, 1)

            # ベクトルから行列
            N1 = torch.mm(n1, n1.view(-1, 2))
            N2 = torch.mm(n2, n2.view(-1, 2))
            N3 = torch.mm(n3, n3.view(-1, 2))
            N4 = torch.mm(n4, n4.view(-1, 2))

            R = c_val[i][0]*(I - N1) + c_val[i][1]*(I - N2) + c_val[i][2]*(I - N3) + c_val[i][3]*(I - N4) # Rの演算
            q = c_val[i][0]*(torch.mm(I - N1, pos_rx1)) + c_val[i][1]*(torch.mm(I - N2, pos_rx2)) + c_val[i][2]*(torch.mm(I - N3, pos_rx3)) + c_val[i][3]*(torch.mm(I - N4, pos_rx4)) # qの演算

            #　Rの擬似逆行列
            R_pinv = torch.linalg.pinv(R)

            # 推定位置
            pos_est = torch.mm(R_pinv, q).view(-1, 2)
            pos_list.append(pos_est[0])
            pos = torch.stack(pos_list)

        return pos
    

'''関数'''

# 角度の基準をx軸に変換する関数   
def change_angle_reference(df_aoa):
    df_aoa['azi_1'] = -df_aoa['azi_1']
    df_aoa['azi_2'] = 90 -df_aoa['azi_2']
    df_aoa['azi_3'] = 180 -df_aoa['azi_3']
    df_aoa['azi_4'] = -90 -df_aoa['azi_4']
    
    return df_aoa

# データフレームにxとyのラベルをつける関数
def set_label(df, label_df):
    for i in np.arange(len(label_df)):
        df.loc[(df['time'] >= label_df.iloc[i]['start_time']) & (df['time'] <= label_df.iloc[i]['end_time']), 'x_label'] = label_df.iloc[i]['x_label'] / 100
        df.loc[(df['time'] >= label_df.iloc[i]['start_time']) & (df['time'] <= label_df.iloc[i]['end_time']), 'y_label'] = label_df.iloc[i]['y_label'] / 100
        
    return df

# 最小二乗法によって送信機の推定位置座標を計算する関数
def least_square(pos_rx, df_aoa):
    # 受信機位置のベクトル化
    pos_rx1 = np.array(pos_rx[0]).reshape(-1, 1)
    pos_rx2 = np.array(pos_rx[1]).reshape(-1, 1)
    pos_rx3 = np.array(pos_rx[2]).reshape(-1, 1)
    pos_rx4 = np.array(pos_rx[3]).reshape(-1, 1)

    #　推定位置
    pos_est = []

    # 最小二乗法
    for i in np.arange(len(df_aoa)):
        I = np.eye(2) # 単位行列

        # 初期化
        R = 0
        q = 0

        # 方向ベクトル（AoA）
        n1 = np.array([math.cos(math.radians(df_aoa['azi_1'][i])), math.sin(math.radians(df_aoa['azi_1'][i]))]).reshape(-1, 1)
        n2 = np.array([math.cos(math.radians(df_aoa['azi_2'][i])), math.sin(math.radians(df_aoa['azi_2'][i]))]).reshape(-1, 1)
        n3 = np.array([math.cos(math.radians(df_aoa['azi_3'][i])), math.sin(math.radians(df_aoa['azi_3'][i]))]).reshape(-1, 1)
        n4 = np.array([math.cos(math.radians(df_aoa['azi_4'][i])), math.sin(math.radians(df_aoa['azi_4'][i]))]).reshape(-1, 1)

        # ベクトルから行列
        N1 = np.dot(n1, n1.reshape(1, -1))
        N2 = np.dot(n2, n2.reshape(1, -1))
        N3 = np.dot(n3, n3.reshape(1, -1))
        N4 = np.dot(n4, n4.reshape(1, -1))

        R = I - N1 + I - N2 + I - N3 + I - N4 # Rの演算
        q = np.dot(I - N1, pos_rx1) + np.dot(I - N2, pos_rx2) + np.dot(I - N3, pos_rx3) + np.dot(I - N4, pos_rx4) # qの演算

        #　Rの擬似逆行列
        R_pinv = np.linalg.pinv(R)

        # 推定位置
        pos = np.dot(R_pinv, q)

        #リストに追加
        pos_est.append(np.dot(R_pinv, q))
        
    return pos_est

# 定数
POS_RX = [[0.0, 3.0], [6.0, 0.0], [12.0, 3.0], [6.0,6.0]]
CHANNEL = 37
START_TIME = 30
END_TIME = 210
```

```python
'''データフレームの生成'''

# カラム
columns = ['time', 'tag_id', 'rss_1', 'azi', 'ele', 'rss_2', 'channel', 'anchor_id']
# データフレーム生成
angle = pd.read_csv('dataset/ieee_dataset/beacons_calibration.txt', sep=',', header=None, names=columns)
# 方位角を取得してカラム名を受信機番号に対応させる
angle = angle[['azi', 'channel', 'anchor_id', 'time']].set_axis(['azi', 'channel', 'anchor_id', 'time'], axis=1).dropna()

# 受信機ごとにデータフレームを分割
df_01 = angle[angle['anchor_id'] == 6501] # 受信機1
df_02 = angle[angle['anchor_id'] == 6502] # 受信機2
df_03 = angle[angle['anchor_id'] == 6503] # 受信機3
df_04 = angle[angle['anchor_id'] == 6504] # 受信機4

# 37チャネルのデータを抽出（37,38,39から選択）
df_01 = df_01[df_01['channel'] == 37]
df_02 = df_02[df_02['channel'] == 37]
df_03 = df_03[df_03['channel'] == 37]
df_04 = df_04[df_04['channel'] == 37]

# タイムスタンプをmsからsに変換
df_01['time'] = (df_01['time']//1000).astype('int')
df_02['time'] = (df_02['time']//1000).astype('int')
df_03['time'] = (df_03['time']//1000).astype('int')
df_04['time'] = (df_04['time']//1000).astype('int')

# 1秒ごとに代表値として中央値をとる
df_01 = df_01.groupby('time').median().astype('int')
df_02 = df_02.groupby('time').median().astype('int')
df_03 = df_03.groupby('time').median().astype('int')
df_04 = df_04.groupby('time').median().astype('int')

# 列名の変更
df_01 = df_01.rename(columns={'azi': 'azi_1'})
df_02 = df_02.rename(columns={'azi': 'azi_2'})
df_03 = df_03.rename(columns={'azi': 'azi_3'})
df_04 = df_04.rename(columns={'azi': 'azi_4'})

# 各データフレームをリストに格納
list_of_series = [df_01['azi_1'], df_02['azi_2'], df_03['azi_3'], df_04['azi_4']]
# 四つの受信機のデータフレームを一つに結合
df = pd.concat(list_of_series, join='inner', axis=1).reset_index(drop=False)  # joinをinnerにすることで、timestampの等しい行を残して結合
# AoAの基準をx軸に変換
df = change_angle_reference(df)
df
```

```python
'''ラベル付け'''

# カラム
columns = ['start_time', 'end_time', 'x_label', 'y_label']
# ラベルのデータフレーム生成
label_df = pd.read_csv('dataset/ieee_dataset/gt_calibration.txt', sep=',', header=None, names=columns)
label_df = label_df[['start_time', 'end_time', 'x_label', 'y_label']].set_axis(['start_time', 'end_time', 'x_label', 'y_label'], axis=1).dropna()

# タイムスタンプを秒単位に変換
label_df['start_time'] = (label_df['start_time']//1000).astype('int')
label_df['end_time'] = (label_df['end_time']//1000).astype('int')

# ラベル付け
df = set_label(df, label_df)
df = df.dropna().reset_index(drop=True)

df
```

```python
'''最小二乗法による推定値の追加'''

data = least_square(POS_RX, df)

x_ls = []
y_ls = []

for i in np.arange(len(df)):
    x_ls.append(round(data[i][0][0], 2))
    y_ls.append(round(data[i][1][0], 2))
    
series_x_ls = pd.Series(x_l, name='x_ls')
series_y_ls = pd.Series(y_l, name='y_ls')

df_ls = pd.concat([series_x_ls,series_y_ls], axis=1)
df = pd.concat([df, df_ls], axis=1)
df
```

```python
'''NN適用前のヒートマップ作成'''

# 誤差の列を追加
df['error_ls'] = np.sqrt((df['x_label'] - df['x_ls'])**2 + (df['y_label'] - df['y_ls'])**2)

#カラーマップ
cmap_ls = np.zeros([7, 17])

xx = [1.2,1.8,2.4,3.0,3.6,4.2,4.8,5.4,6.0,6.6,7.2,7.8,8.4,9.0,9.6,10.2,10.8]
yy = [4.8,4.2,3.6,3.0,2.4,1.8,1.2]

for i in np.arange(7):
    for n in np.arange(17):
        cmap_ls[i][n] = df[(df['x_label'] == xx[n]) & (df['y_label'] == yy[i])]['error_ls'].mean()
        
df_ls = pd.DataFrame(data=cmap_ls, index=yy, columns=xx)
plt.figure(figsize=(17, 7))
sns.heatmap(df_ls, cmap='OrRd',vmax=3, vmin=-0)
plt.savefig("figure/heatmap_ls.png")
```

```python
'''データセットの準備'''

# 訓練データと検証&テストデータに分割
train, val_test = train_test_split(df, test_size=0.5, random_state=2023)
    
# 検証データとテストデータに分割
val, test = train_test_split(val_test, test_size=0.5, random_state=2023)

# 特徴量の名前
features = ['azi_1', 'azi_2', 'azi_3', 'azi_4']

# ラベルの名称
label = ['x_label', 'y_label']

# データセットの作成
train_dataset = MyDataset(train, features, label)
val_dataset = MyDataset(val, features, label)
test_dataset = MyDataset(test, features, label)

#訓練データ，検証データ，テストデータのDataLoaderを生成
train_dataloader = DataLoader(train_dataset,
                              batch_size=2000,
                              shuffle=True,
                              drop_last=True)

val_dataloader = DataLoader(val_dataset,
                            batch_size=100,
                            shuffle=False)

# test_dataloaderは上から4ごとにposA, posB, posC, posDとなっている
test_dataloader = DataLoader(test_dataset,
                             batch_size=100,
                             shuffle=False)
```

```python
'''NNモデルの準備'''
model = Net().to(device)
```

```python
'''NNの訓練準備'''

# 損失関数
criterion = nn.MSELoss()
# 勾配アルゴリズム
optimizer = optimizers.Adam(model.parameters(),
                            lr=0.0001, # 学習率
                            betas=(0.9, 0.999),
                            amsgrad=True)

# 損失の計算
def compute_loss(t, y):
    return criterion(y, t)

# 訓練用関数
def train_step(x, t):
    model.train()
    preds = model(x)
    loss = compute_loss(t, preds)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    return loss, preds

# 検証用関数
def val_step(x, t):
    model.eval() # モデルを推論モードへ切り替え
    preds = model(x)
    loss = criterion(preds, t)

    return loss, preds

# エポック数
epochs = 100

# 損失格納用リスト
hist = {'loss': [], 'val_loss': []}
```

```python
'''NNモデルの訓練'''

for epoch in range(epochs):
    train_loss = 0.
    val_loss = 0.

    for (x, t) in train_dataloader:
        x, t = x.to(device), t.to(device).view(-1, 2)
        loss, preds = train_step(x, t)
        train_loss += loss.item()

    train_loss /= len(train_dataloader)

    for (x, t) in val_dataloader:
        x, t = x.to(device), t.to(device).view(-1, 2)
        loss, preds = val_step(x, t)
        val_loss += loss.item()

    val_loss /= len(val_dataloader)

    hist['loss'].append(train_loss)
    hist['val_loss'].append(val_loss)

    print('epoch: {}, loss: {:.3}, val_loss: {:.3}'.format(
              epoch+1,
              train_loss,
              val_loss
          ))
```

```python
'''学習状況の可視化（val_lossの推移）'''

train_loss = hist['loss']
val_loss = hist['val_loss']

fig = plt.figure()
plt.plot(train_loss, color='red', linewidth=1, label="train")
plt.plot(range(len(val_loss)), val_loss, color='blue', linewidth=1, label="validation")
plt.xlabel('epochs')
plt.ylabel('loss')
plt.title('val_loss')
plt.legend()
plt.savefig("figure/train_val_loss_2.png")
plt.show()
```

```python
'''NNモデルの評価'''

# 評価用関数
def test_step(x, t):
    return val_step(x, t)

# 測位結果格納用リスト
nn = []

for (x, t) in test_dataloader:
    x, t = x.to(device), t.to(device).view(-1, 2)
    loss, preds = test_step(x, t)

    p_t = torch.cat((preds, t), dim=1)

    nn.append(p_t)
```

```python
'''NNによる測位結果のデータフレーム作成'''
nn = torch.cat(nn, dim=0)
nn = nn.detach().numpy().copy()
nn_columns = ['x_est', 'y_est', 'x_lab', 'y_lab']
nn = pd.DataFrame(nn, columns=nn_columns)
nn
```

```python
'''誤差の列を追加'''
nn['error'] = np.sqrt((nn['x_lab'] - nn['x_est'])**2 + (nn['y_lab'] - nn['y_est'])**2)
nn
```

```python
'''測位誤差のヒートマップ作成'''

# 地点ごとの測位誤差格納用配列
cmap = np.zeros([7, 17])

# x軸とy軸
xx = [1.2,1.8,2.4,3.0,3.6,4.2,4.8,5.4,6.0,6.6,7.2,7.8,8.4,9.0,9.6,10.2,10.8]
yy = [4.8,4.2,3.6,3.0,2.4,1.8,1.2]

# 測位誤差を計算し配列に格納
for i in np.arange(7):
    for n in np.arange(17):
        cmap[i][n] = nn[(nn['x_lab'] == xx[n]) & (nn['y_lab'] == yy[i])]['error'].mean()

# ヒートマップ用のデータフレーム
df_c = pd.DataFrame(data=cmap, index=yy, columns=xx)
# ヒートマップの作成
plt.figure(figsize=(17, 7))
sns.heatmap(df_c, cmap='OrRd',vmax=3.0, vmin=-0)
plt.savefig("figure/heatmap_nn.png")
```
