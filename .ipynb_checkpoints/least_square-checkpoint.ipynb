```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import seaborn as sns

# 図のフォント指定
plt.rcParams['font.family'] = 'Times New Roman'
sns.set_style({'font.family':'serif', 'font.serif':['Times New Roman']})

'''メソッド'''

# 生データ→データフレーム
def preprocessing_aoa(list_of_path, channel, start_time, end_time): # RX1~RX4のパスのリスト、チャネル、開始時間、終了時間が引数
    
    # 各受信機のdfを格納
    angles = []

    # 各受信機ごとに前処理を実施
    for i in np.arange(len(list_of_path)):
        # カラム
        columns = ["instanceID", "RSSI", "Azimuth", "Elevation", "Reserved parameter", "advertising channel", "Anchor ID", "Userdefinedstrings", "Timestamp"]
        # データフレーム生成
        angle = pd.read_csv(list_of_path[i], sep=',', header=None, names=columns)
        # 方位角を取得してカラム名を受信機番号に対応させる
        angle = angle[['Azimuth', 'advertising channel', 'Timestamp']].set_axis(['azi_%d' % (i+1), 'channel', 'timestamp'], axis=1).dropna()
        # 60秒から18秒までを抽出
        angle = angle[angle.timestamp > start_time*1000]
        angle = angle[angle.timestamp < end_time*1000]
        # 指定チャネルのデータを抽出
        angle = angle[angle['channel'].isin([channel])]
        # timestampを秒単位に統一
        angle['timestamp'] = (angle['timestamp']//1000).astype('int')
        # 1秒ごとに中央値をとる
        angle= angle.groupby('timestamp').median().astype('int')
        
        angles.append(angle)  # リストに格納
        
    list_of_series = [angles[0]['azi_1'], angles[1]['azi_2'], angles[2]['azi_3'], angles[3]['azi_4']]  # Seriesに分解して並び替える
    
    df = pd.concat(list_of_series, join='inner', axis=1).reset_index(drop=True)  # joinをinnerにすることで、timestampの等しい行を残して結合

    return df

# 角度の基準をx軸に変換する
def change_angle_reference(df_aoa):
    df_aoa['azi_1'] = -df_aoa['azi_1']
    df_aoa['azi_2'] = -90 -df_aoa['azi_2']
    df_aoa['azi_3'] = 180 -df_aoa['azi_3']
    df_aoa['azi_4'] = 90 -df_aoa['azi_4']
    
    return df_aoa

# 最小二乗法によって送信機の推定位置座標を計算
def least_square(pos_rx, df_aoa):
    # 受信機位置のベクトル化
    pos_rx1 = np.array(pos_rx[0]).reshape(-1, 1)
    pos_rx2 = np.array(pos_rx[1]).reshape(-1, 1)
    pos_rx3 = np.array(pos_rx[2]).reshape(-1, 1)
    pos_rx4 = np.array(pos_rx[3]).reshape(-1, 1)

    #　推定位置
    pos_est = []

    # 最小二乗法
    for i in np.arange(len(df_aoa)):
        I = np.eye(2) # 単位行列

        # 初期化
        R = 0
        q = 0

        # 方向ベクトル（AoA）
        n1 = np.array([math.cos(math.radians(df_aoa['azi_1'][i])), math.sin(math.radians(df_aoa['azi_1'][i]))]).reshape(-1, 1)
        n2 = np.array([math.cos(math.radians(df_aoa['azi_2'][i])), math.sin(math.radians(df_aoa['azi_2'][i]))]).reshape(-1, 1)
        n3 = np.array([math.cos(math.radians(df_aoa['azi_3'][i])), math.sin(math.radians(df_aoa['azi_3'][i]))]).reshape(-1, 1)
        n4 = np.array([math.cos(math.radians(df_aoa['azi_4'][i])), math.sin(math.radians(df_aoa['azi_4'][i]))]).reshape(-1, 1)

        # ベクトルから行列
        N1 = np.dot(n1, n1.reshape(1, -1))
        N2 = np.dot(n2, n2.reshape(1, -1))
        N3 = np.dot(n3, n3.reshape(1, -1))
        N4 = np.dot(n4, n4.reshape(1, -1))

        R = I - N1 + I - N2 + I - N3 + I - N4 # Rの演算
        q = np.dot(I - N1, pos_rx1) + np.dot(I - N2, pos_rx2) + np.dot(I - N3, pos_rx3) + np.dot(I - N4, pos_rx4) # qの演算

        #　Rの擬似逆行列
        R_pinv = np.linalg.pinv(R)

        # 推定位置
        pos = np.dot(R_pinv, q)

        #リストに追加
        pos_est.append(np.dot(R_pinv, q))
        
    return pos_est

# 水平距離誤差の計算
def calculate_distance_error(pos_est, ground_truth):
    # 誤差の総和
    error = []

    # 水平距離誤差の計算
    for i in np.arange(len(pos_est)):
        error.append(math.sqrt((ground_truth[0]-pos_est[i][0])**2 +(ground_truth[1]-pos_est[i][1])**2))
    
    return error

def calculate_ave_distance_error(pos_est, ground_truth):
    # 誤差の総和
    total_error = 0

    # 水平距離誤差の計算
    for i in np.arange(len(pos_est)):
        total_error += math.sqrt((ground_truth[0]-pos_est[i][0])**2 +(ground_truth[1]-pos_est[i][1])**2)

    # 誤差の平均
    ave_error = total_error/len(pos_est)
    
    return ave_error

# AoAのグランドトゥルースの計算
def calculate_aoa(pos_tx):
    
    # 受信機位置座標（設置位置に応じて適宜変更する）
    pos_rx1 = np.array([0,2])
    pos_rx2 = np.array([2,4])
    pos_rx3 = np.array([4,2])
    pos_rx4 = np.array([2,0])
    
    # x軸方向の単位ベクトル
    vec_nx = np.array([1,0])
    
    """ 方位角の計算 """
    
    # 各受信機と送信機とのベクトル
    vec_1t = pos_tx - pos_rx1
    vec_2t = pos_tx - pos_rx2
    vec_3t = pos_tx - pos_rx3
    vec_4t = pos_tx - pos_rx4
    
    # ノルムの算出
    len_vec_1t = round(np.linalg.norm(vec_1t), 5)
    len_vec_2t = round(np.linalg.norm(vec_2t), 5)
    len_vec_3t = round(np.linalg.norm(vec_3t), 5)
    len_vec_4t = round(np.linalg.norm(vec_4t), 5)
    
    # 内積の算出
    inner_product_nx_1t = np.inner(vec_nx, vec_1t)
    inner_product_nx_2t = np.inner(vec_nx, vec_2t)
    inner_product_nx_3t = np.inner(vec_nx, vec_3t)
    inner_product_nx_4t = np.inner(vec_nx, vec_4t)
    
    # コサインの計算
    cos_a1 = round(inner_product_nx_1t / (1 * len_vec_1t), 5)
    cos_a2 = round(inner_product_nx_2t / (1 * len_vec_2t), 5)
    cos_a3 = round(inner_product_nx_3t / (1 * len_vec_3t), 5)
    cos_a4 = round(inner_product_nx_4t / (1 * len_vec_4t), 5)
    
    # 方位角（ラジアン）の計算
    rad_a1 = round(np.arccos(cos_a1), 5)
    rad_a2 = round(np.arccos(cos_a2), 5)
    rad_a3 = round(np.arccos(cos_a3), 5)
    rad_a4 = round(np.arccos(cos_a4), 5)
    
    # ラジアン→度
    deg_a1 = round(np.rad2deg(rad_a1), 5)
    deg_a2 = round(np.rad2deg(rad_a2), 5)
    deg_a3 = round(np.rad2deg(rad_a3), 5)
    deg_a4 = round(np.rad2deg(rad_a4), 5)
    
    # x軸を基準としたAoAに変換（受信機や送信機の位置によってこの処理は適宜書き換える必要あり）
    if pos_tx[1] >= 2:
        azi_1 = round(deg_a1, 2)
        azi_3 = round(deg_a3, 2)
    else:
        azi_1 = round(-deg_a1, 2)
        azi_3 = round(360-deg_a3, 2)
        
    azi_2 = round(-deg_a2, 2)
    azi_4 = round(deg_a4, 2)
    
    return [azi_1, azi_2, azi_3, azi_4]

'''定数'''

# 受信機位置
POS_RX = [[0.0, 2.0], [2.0, 4.0], [4.0, 2.0], [2.0,0.0]]
```

```python
'''pos1'''

# パスの指定
path = ['data/20231004/pos1_rx1.log', 'data/20231004/pos1_rx2.log', 'data/20231004/pos1_rx3.log', 'data/20231004/pos1_rx4.log']

# データフレーム生成
df_aoa_pos1_37 = preprocessing_aoa(path, 37, 60, 180)

# AoAの基準をx軸に変更
df_aoa_pos1_37 = change_angle_reference(df_aoa_pos1_37)

# 推定位置座標の算出
pos_est = least_square(POS_RX, df_aoa_pos1_37)

# 推定位置座標とグランドトゥルースの水平距離誤差
error_pos1 = calculate_distance_error(pos_est, [1.5, 2.5])

# 推定位置座標とグランドトゥルースの水平距離誤差の平均
ave_error_pos1 = calculate_ave_distance_error(pos_est, [1.5, 2.5])

# 推定座標をxyに分割
pos_est_x = []
pos_est_y = []

for i in np.arange(len(pos_est)):
    pos_est_x.append(pos_est[i][0])
    pos_est_y.append(pos_est[i][1])
    
# 散布図
plt.figure(figsize=(6, 6))
plt.grid()
plt.scatter(pos_est_x, pos_est_y, c="blue", label="least_square")
plt.scatter(1.5, 2.5, c="lime", s=100, label="True")
plt.xlabel('x [m]', fontsize=35)
plt.ylabel('y [m]', fontsize=35)
plt.xticks(np.arange(0, 5.0), fontsize= 30)
plt.yticks(np.arange(0, 5.0), fontsize= 30)
plt.legend(fontsize=10)
plt.tight_layout()
```
